```python

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python

stocks = {
    "TCS": "TCS.NS",
    "RELIANCE": "RELIANCE.NS",
    "HDFCBANK": "HDFCBANK.NS",
    "INFY": "INFY.NS",
    "ICICIBANK": "ICICIBANK.NS",
    "SBIN": "SBIN.NS",
    "ITC": "ITC.NS",
    "HINDUNILVR": "HINDUNILVR.NS",
    "LT": "LT.NS",
    "AXISBANK": "AXISBANK.NS"
}
```

```python

data = {}

for name, symbol in stocks.items():
    df = yf.download(symbol, period="2y", progress=False)
    df = df[["Close"]]                # keep only Close price
    df.reset_index(inplace=True)      # bring Date as a column
    data[name] = df
```

```python
TCS = data["TCS"][["Date", "Close"]].copy()
RELIANCE = data["RELIANCE"][["Date", "Close"]].copy()
HDFCBANK = data["HDFCBANK"][["Date", "Close"]].copy()
INFY = data["INFY"][["Date", "Close"]].copy()
ICICIBANK = data["ICICIBANK"][["Date", "Close"]].copy()
SBIN = data["SBIN"][["Date", "Close"]].copy()
ITC = data["ITC"][["Date", "Close"]].copy()
HINDUNILVR = data["HINDUNILVR"][["Date", "Close"]].copy()
LT = data["LT"][["Date", "Close"]].copy()
AXISBANK = data["AXISBANK"][["Date", "Close"]].copy()
```

```python
for df in [TCS, RELIANCE, HDFCBANK, INFY, ICICIBANK, SBIN, ITC, HINDUNILVR, LT, AXISBANK]:
    df["SMA_50"] = df["Close"].rolling(window=50).mean()
    df["SMA_200"] = df["Close"].rolling(window=200).mean()
```

```python

# TCS
plt.figure(figsize=(10,5))
plt.plot(TCS["Date"], TCS["Close"], label="Close Price")
plt.plot(TCS["Date"], TCS["SMA_50"], label="SMA 50")
plt.plot(TCS["Date"], TCS["SMA_200"], label="SMA 200")
plt.title("TCS Stock Price with SMA")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()

# RELIANCE
plt.figure(figsize=(10,5))
plt.plot(RELIANCE["Date"], RELIANCE["Close"], label="Close Price")
plt.plot(RELIANCE["Date"], RELIANCE["SMA_50"], label="SMA 50")
plt.plot(RELIANCE["Date"], RELIANCE["SMA_200"], label="SMA 200")
plt.title("RELIANCE Stock Price with SMA")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()

# HDFCBANK
plt.figure(figsize=(10,5))
plt.plot(HDFCBANK["Date"], HDFCBANK["Close"], label="Close Price")
plt.plot(HDFCBANK["Date"], HDFCBANK["SMA_50"], label="SMA 50")
plt.plot(HDFCBANK["Date"], HDFCBANK["SMA_200"], label="SMA 200")
plt.title("HDFCBANK Stock Price with SMA")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()

# INFY
plt.figure(figsize=(10,5))
plt.plot(INFY["Date"], INFY["Close"], label="Close Price")
plt.plot(INFY["Date"], INFY["SMA_50"], label="SMA 50")
plt.plot(INFY["Date"], INFY["SMA_200"], label="SMA 200")
plt.title("INFY Stock Price with SMA")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()

# ICICIBANK
plt.figure(figsize=(10,5))
plt.plot(ICICIBANK["Date"], ICICIBANK["Close"], label="Close Price")
plt.plot(ICICIBANK["Date"], ICICIBANK["SMA_50"], label="SMA 50")
plt.plot(ICICIBANK["Date"], ICICIBANK["SMA_200"], label="SMA 200")
plt.title("ICICIBANK Stock Price with SMA")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()

# SBIN
plt.figure(figsize=(10,5))
plt.plot(SBIN["Date"], SBIN["Close"], label="Close Price")
plt.plot(SBIN["Date"], SBIN["SMA_50"], label="SMA 50")
plt.plot(SBIN["Date"], SBIN["SMA_200"], label="SMA 200")
plt.title("SBIN Stock Price with SMA")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()

# ITC
plt.figure(figsize=(10,5))
plt.plot(ITC["Date"], ITC["Close"], label="Close Price")
plt.plot(ITC["Date"], ITC["SMA_50"], label="SMA 50")
plt.plot(ITC["Date"], ITC["SMA_200"], label="SMA 200")
plt.title("ITC Stock Price with SMA")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()

# HINDUNILVR
plt.figure(figsize=(10,5))
plt.plot(HINDUNILVR["Date"], HINDUNILVR["Close"], label="Close Price")
plt.plot(HINDUNILVR["Date"], HINDUNILVR["SMA_50"], label="SMA 50")
plt.plot(HINDUNILVR["Date"], HINDUNILVR["SMA_200"], label="SMA 200")
plt.title("HINDUNILVR Stock Price with SMA")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()

# LT
plt.figure(figsize=(10,5))
plt.plot(LT["Date"], LT["Close"], label="Close Price")
plt.plot(LT["Date"], LT["SMA_50"], label="SMA 50")
plt.plot(LT["Date"], LT["SMA_200"], label="SMA 200")
plt.title("LT Stock Price with SMA")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()

# AXISBANK
plt.figure(figsize=(10,5))
plt.plot(AXISBANK["Date"], AXISBANK["Close"], label="Close Price")
plt.plot(AXISBANK["Date"], AXISBANK["SMA_50"], label="SMA 50")
plt.plot(AXISBANK["Date"], AXISBANK["SMA_200"], label="SMA 200")
plt.title("AXISBANK Stock Price with SMA")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
```
